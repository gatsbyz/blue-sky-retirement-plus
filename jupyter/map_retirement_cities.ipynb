```python
import folium
import json

import sys
sys.path.insert(0, '..')
import src.fileio as fileio


f = open('../static/location-data.json')
rawData = json.load(f)
f.close()

retirement_cities = fileio.load_csv("../cache/retirement-cities.csv")
print (retirement_cities)

cities = rawData
```

```python
my_map3 = folium.Map(location=[13/133932434766733, 16.10398729508073],
    zoom_start=2)
```

```python

for city in retirement_cities:
    mk = folium.Marker(rawData[city]['lat-long'])
    pp = folium.Popup(city)
    ic = folium.Icon(rawData[city]['color'])

    mk.add_child(ic)
    mk.add_child(pp)
    my_map3.add_child(mk)


my_map3.save("../Interactive_retirement_map.html")

my_map3
```
